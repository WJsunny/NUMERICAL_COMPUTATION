```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
g = 9.81
c = 12.5
m = 68.1
    
def findVelocity(dx, tfn):
    dx = float(dx)
    tfn = float(tfn)
    ti = 0.0
    vi =0.0
    x = np.arange(0.0,tfn,dx)
    y = np.array([0.0])
    for ti in x:
        ti1 = ti + dx
        vi1 = vi + (g - (c/m)*vi)*(ti1 - ti)
        y = np.append(y,vi1)
        vi = vi1
    x = np.arange(0.0,tfn+dx,dx)
    return x, y

def findV(tfn):
    tfn = float(tfn)
    t = np.arange(0,tfn,2)
    v = ((g*m)/c)*(1-(np.exp(-(c/m)*t)))
    return t, v
```

```python
x1, y1 = findVelocity(2,30)
x2, y2 = findV(30)
fig1 = plt.figure()
plt.plot(x1, y1,'*-', label = "Numerical Solution")
#fig2 = plt.figure()
plt.plot(x2, y2, label = "Analytical Solution")
plt.legend()
plt.grid()
plt.title("Graph")
plt.xlabel('Time(s)')
plt.ylabel('Velocity(v/s)')
```
