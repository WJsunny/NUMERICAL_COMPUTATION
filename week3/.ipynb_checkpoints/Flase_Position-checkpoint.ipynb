```python
import numpy as np
```

```python
def FlasePosition(xl, xu, err, fnc):
    xrn = 0
    i = 0
    il = 0
    iu = 0
    er = err +10
    while not er <= err:
        xr = xrn
        fxl = fnc(xl)
        fxu = fnc(xu)
        xrn = xu-(fxu*(xl-xu))/(fxl - fxu)
        fxr = fnc(xrn)
        if (fxl * fxr <0):
            xu = xrn
            iu = 0
            il += 1
            if (il >= 2):
                fxl /= 2
        elif (fxl * fxr >0):
            xl = xrn
            il = 0
            iu += 1
            if (iu >= 2):
                fxu /= 2
        else:
            print('i = {:3d}, x = {:6.3f}, error = {:.3f}%'.format(i,xrn,er))
            break
        if not xrn == 0:
            er = np.abs((xrn - xr)/xrn) * 100
        print('i = {:3d}, x = {:6.3f}, error = {:.3f}%'.format(i,xrn,er))
        i += 1

def fc(c):
    g= 9.81
    m = 68.1
    v = 40.0
    t = 10.0
    y = ((g*m)/c)*(1-np.exp(-(c/m)*t)) - 40.0
    return y
```

```python
FlasePosition(-20, 25, 0.001, fc)
```
